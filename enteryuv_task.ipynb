```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataframe is named final_df1
# Replace 'final_df1' with your actual dataframe name

# Select relevant features for clustering
features_for_clustering = ['QUANTITY', 'IMPORT VALUE FOB']

# Drop rows with missing values for selected features
data_for_clustering = final_df1[features_for_clustering].dropna()

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_for_clustering)

# Determine the optimal number of clusters using the elbow method
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot the elbow method
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()

# Based on the elbow method, choose the optimal K (number of clusters)
optimal_k = 3  # You may adjust this based on the plot

# Apply K-means clustering
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
final_df1['Cluster'] = kmeans.fit_predict(scaled_data)

# Visualize the clusters
sns.scatterplot(x='QUANTITY', y='IMPORT VALUE FOB', hue='Cluster', data=final_df1, palette='viridis')
plt.title('Clusters of Rice Export Data')
plt.xlabel('Quantity')
plt.ylabel('Import Value FOB')
plt.show()
```

```python
from sklearn.ensemble import IsolationForest

features_for_anomaly = ['QUANTITY', 'IMPORT VALUE FOB']

# Drop rows with missing values for selected features
data_for_anomaly = final_df1[features_for_anomaly].dropna()

# Standardize the data
scaler = StandardScaler()
scaled_data_anomaly = scaler.fit_transform(data_for_anomaly)

# Apply Isolation Forest
isolation_forest = IsolationForest(contamination=0.05, random_state=42)
final_df1['Anomaly'] = isolation_forest.fit_predict(scaled_data_anomaly)

# Visualize anomalies
sns.scatterplot(x='QUANTITY', y='IMPORT VALUE FOB', hue='Anomaly', data=final_df1, palette='viridis')
plt.title('Anomalies Detected in Rice Export Data')
plt.xlabel('Quantity')
plt.ylabel('Import Value FOB')
plt.show()
```
